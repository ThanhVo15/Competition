## 2. Production

```python

production_data = df[df['Attribute_Description'].isin(['Arabica Production', 'Robusta Production', 'Other Production', 'Production'])]
production_data = production_data.groupby(['Calendar_Year', 'Attribute_Description'])['Value'].sum().unstack()

# Plotting stacked area chart
plt.figure(figsize=(12, 6))
plt.stackplot(production_data.index, production_data[['Arabica Production', 'Robusta Production', 'Other Production', 'Production']].T, labels=['Arabica Production', 'Robusta Production', 'Other Production', 'Total Production'], alpha=0.8)
plt.title('Production Distribution Over Time')
plt.xlabel('Calendar Year')
plt.ylabel('Production Value')
plt.legend()
plt.show()
```

```python
# Select relevant columns and group by 'Market_Year'
production_data = df[df['Attribute_Description'].isin(['Arabica Production', 'Robusta Production', 'Bean Imports', 'Bean Exports'])]
production_data = production_data.groupby(['Calendar_Year', 'Attribute_Description'])['Value'].sum().unstack()

# Plotting line chart for all variables
plt.figure(figsize=(12, 6))

plt.plot(production_data.index, production_data['Arabica Production'], label='Arabica Production', marker='o')
plt.plot(production_data.index, production_data['Robusta Production'], label='Robusta Production', marker='o')
plt.plot(production_data.index, production_data['Bean Imports'], label='Bean Imports', marker='o')
plt.plot(production_data.index, production_data['Bean Exports'], label='Bean Exports', marker='o', color='black')

plt.title('Production Trends Over Time')
plt.xlabel('Calendar Year')
plt.ylabel('Production Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Select relevant columns and filter by a specific country
country_data = df[(df['Country_Name'] == 'Brazil') & df['Attribute_Description'].isin(['Arabica Production', 'Robusta Production', 'Other Production', 'Total Production'])]
country_data = country_data.groupby(['Calendar_Year', 'Attribute_Description'])['Value'].sum().unstack()

# Create a new column for 'Production' which is the sum of the three types
country_data['Production'] = country_data[['Arabica Production', 'Robusta Production', 'Other Production']].sum(axis=1)

# Plotting stacked area chart
plt.figure(figsize=(12, 6))
plt.stackplot(country_data.index, country_data[['Arabica Production', 'Robusta Production', 'Other Production', 'Production']].T, labels=['Arabica Production', 'Robusta Production', 'Other Production', 'Total Production'], alpha=0.8)
plt.title('Production Distribution Over Time in Brazil')
plt.xlabel('Calendar Year')
plt.ylabel('Production Value')
plt.legend()
plt.show()
```

# III. Plotting base on Categorical Country Name

- **Tỷ lệ nhập khẩu trên tiêu thụ (Import/Consumption ratio):** Tỷ lệ này cho biết lượng cà phê nhập khẩu chiếm bao nhiêu phần trăm tổng lượng cà phê tiêu thụ của một nước. Nếu tỷ lệ này lớn hơn 1, thì nước đó được coi là nước thuần nhập khẩu.
- **Tỷ lệ xuất khẩu trên sản lượng (Export/Production ratio):** Tỷ lệ này cho biết lượng cà phê xuất khẩu chiếm bao nhiêu phần trăm tổng lượng cà phê sản xuất của một nước. Nếu tỷ lệ này lớn hơn 1, thì nước đó được coi là nước thuần xuất khẩu.

```python
df['Country_Name'].unique().tolist()
```

```python
# Định nghĩa từ điển phân loại quốc gia theo châu lục
continent_dict = {
    'Africa': ['Algeria', 'Angola', 'Benin', 'Burundi', 'Cameroon', 'Central African Republic', 'Equatorial Guinea',
               'Ethiopia', 'Gabon', 'Ghana', 'Guinea', 'Ivory Coast', 'Kenya', 'Liberia', 'Madagascar', 'Malawi',
               'Morocco', 'Nigeria', 'Rwanda', 'Senegal', 'Sierra Leone', 'South Africa', 'Tanzania', 'Togo', 'Uganda',
               'Zambia', 'Zimbabwe'],
    'Asia': ['Armenia', 'China', 'India', 'Indonesia', 'Iran', 'Japan', 'Jordan', 'Kazakhstan', 'Laos', 'Malaysia',
             'Philippines', 'Russia', 'Saudi Arabia', 'Singapore', 'Taiwan', 'Thailand', 'Turkey', 'Vietnam', 'Yemen',
             'Yemen (Sanaa)'],
    'Europe': ['Albania', 'Bosnia and Herzegovina', 'Croatia', 'Georgia', 'Montenegro', 'North Macedonia', 'Norway',
               'Serbia', 'Switzerland', 'Kosovo'],
    'North America': ['Canada', 'United States', 'Mexico'],
    'Oceania': ['Australia', 'New Caledonia', 'New Zealand', 'Papua New Guinea'],
    'South America': ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic',
                      'Ecuador', 'El Salvador', 'Guatemala', 'Guyana', 'Haiti', 'Honduras', 'Jamaica', 'Panama',
                      'Paraguay', 'Peru', 'Trinidad and Tobago', 'Uruguay', 'Venezuela']
}

# Áp dụng phân loại quốc gia theo châu lục vào DataFrame
df['Continent'] = df['Country_Name'].map(
    {country: continent for continent, countries in continent_dict.items() for country in countries}
)

# Lọc dữ liệu để chỉ giữ lại thông tin liên quan đến sản xuất, xuất khẩu và nhập khẩu
relevant_attributes = ['Exports', 'Bean Imports', 'Production', 'Rst,Ground Dom. Consum', 'Total Supply']
filtered_data = df[df['Attribute_Description'].isin(relevant_attributes)]

# Tính tổng sản xuất, xuất khẩu và nhập khẩu cho mỗi châu lục
continent_aggregated = filtered_data.groupby(['Continent', 'Calendar_Year', 'Attribute_Description']).sum(numeric_only=True).reset_index()

# Tạo lại biểu đồ đường cho mỗi châu lục
plt.figure(figsize=(15, 8))

# Vẽ biểu đồ xu hướng cho nhập khẩu
sns.lineplot(data=continent_aggregated[continent_aggregated['Attribute_Description'] == 'Bean Imports'],
             x='Calendar_Year', y='Value', hue='Continent', marker='o')

# Vẽ biểu đồ xu hướng cho xuất khẩu
sns.lineplot(data=continent_aggregated[continent_aggregated['Attribute_Description'] == 'Exports'],
             x='Calendar_Year', y='Value', hue='Continent', marker='o', linestyle='--')

# Vẽ biểu đồ xu hướng cho Domestic Consumption
sns.lineplot(data=continent_aggregated[continent_aggregated['Attribute_Description'] == 'Rst,Ground Dom. Consum'],
             x='Calendar_Year', y='Value', hue='Continent', marker='o')

plt.title('Import and Export Trends of Coffee Beans by Continent Over Time')
plt.xlabel('Year')
plt.ylabel('Value (1000 60 KG BAGS)')
plt.legend(title='Continent / Type', loc='upper left')
plt.grid(True)
plt.show()
```
