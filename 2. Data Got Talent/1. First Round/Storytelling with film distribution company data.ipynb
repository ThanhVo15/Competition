# MÔ TẢ VẤN ĐỀ
---

Rạp chiếu phim hiện nay đang đối mặt với những thách thức lớn liên quan đến doanh thu, khách hàng, và cách tiếp cận bán chéo sản phẩm một cách hiệu quả. Việc phân tích xem khách hàng trong khoảng thời gian nhất định có quan tâm đến sản phẩm mới mà rạp cung cấp hay không sẽ rất hữu ích cho việc hoạch định chiến lược kinh doanh.

Một rạp chiếu phim không chỉ kiếm doanh thu từ việc bán vé xem phim mà còn từ việc bán thức ăn nhanh, nước uống và các sản phẩm bán chéo khác như quà lưu niệm liên quan đến phim. Ví dụ, một khách hàng có thể chi trả 150.000 đồng cho một vé xem phim và thêm 100.000 đồng cho bắp rang và nước uống. Tuy nhiên, không phải lúc nào khách hàng cũng sẵn lòng chi trả cho các sản phẩm phụ trợ này. Sự quan tâm của họ có thể phụ thuộc vào nhiều yếu tố như sở thích cá nhân, thu nhập, và trải nghiệm trước đó tại rạp.

Nếu rạp chiếu phim biết được khách hàng nào có khả năng cao sẽ mua sản phẩm bán chéo, rạp có thể tối ưu hóa chiến lược tiếp thị của mình, từ việc gửi ưu đãi đặc biệt qua email hoặc tin nhắn đến việc tùy chỉnh quảng cáo trên mạng xã hội để thu hút sự chú ý của họ. Điều này không chỉ giúp tăng doanh thu từ bán hàng chéo mà còn cải thiện trải nghiệm của khách hàng tại rạp, khi họ cảm thấy được cá nhân hóa và chăm sóc tốt hơn.

Để đưa ra những nhận định này, rạp chiếu phim cần phải phân tích dữ liệu về hành vi của khách hàng trong quá khứ, bao gồm lịch sử mua vé và mua hàng phụ trợ, sở thích xem phim, cũng như thông tin demography như tuổi, giới tính, và nơi sinh sống. Sử dụng những thông tin này, bài phân tích có thể giúp rạp chiếu phim hiểu rõ hơn về khách hàng của mình và phục vụ họ một cách hiệu quả hơn.

# THÔNG TIN THUỘC TÍNH
---

1. Sheet "Customer"
- customerid: Mã định danh duy nhất của khách hàng.
- DOB: Ngày sinh của khách hàng, có thể được lưu trữ dưới dạng số (ví dụ: format Excel date).
- gender: Giới tính của khách hàng, ví dụ Nam hoặc Nữ.
- address: Địa chỉ của khách hàng.
- Website: Thông tin liên lạc qua website hoặc một mã định danh liên quan đến khách hàng trên website của rạp.
- job: Nghề nghiệp của khách hàng.
- industry: Ngành nghề mà khách hàng đang làm việc.
2. Sheet "Ticket"
- orderid: Mã định danh duy nhất cho mỗi giao dịch mua vé.
- cashier: Mã nhân viên hoặc tên nhân viên bán vé.
- saledate: Ngày và thời gian bán vé.
- total: Tổng giá trị của giao dịch (có thể bao gồm vé và các sản phẩm khác như popcorn).
- customerid: Mã định danh của khách hàng mua vé.
- ticketcode: Mã vé, có thể là mã duy nhất cho mỗi vé.
- date: Ngày xem phim.
- time: Thời gian bắt đầu chiếu phim.
- slot: Mã ghế hoặc khu vực ghế trong rạp.
- room: Số phòng chiếu phim.
- film: Tên của bộ phim.
- slot type: Loại ghế, ví dụ Đơn cho ghế đơn hoặc có thể là loại ghế VIP, thường...
- ticket type: Loại vé, ví dụ Thành viên cho khách hàng là thành viên hoặc các loại vé khuyến mãi, vé thông thường...
- ticket price: Giá của một vé.
- popcorn: Thông tin về việc khách hàng có mua popcorn hay không.
3. Sheet "Film"
- show_id: Mã định danh duy nhất cho mỗi bộ phim.
- title: Tên của bộ phim.
- director: Đạo diễn của bộ phim.
- cast: Danh sách diễn viên tham gia.
- country: Quốc gia sản xuất phim.
- release_year: Năm phát hành phim.
- rating: Đánh giá phim, có thể là độ tuổi phù hợp để xem phim hoặc xếp hạng chất lượng.
- duration: Thời lượng phim, thường được tính bằng phút.
- listed_in: Thể loại của bộ phim, ví dụ như hành động, phiêu lưu, gia đình, khoa học viễn tưởng...
- description: Mô tả ngắn gọn về nội dung của bộ phim.

Bộ dữ liệu về rạp chiếu phim mà chúng tôi đang khám phá bao gồm thông tin chi tiết từ ba sheet khác nhau: thông tin khách hàng, vé đã bán, và thông tin về các bộ phim được chiếu. Tóm lại, bộ dữ liệu này cung cấp một cái nhìn toàn diện về hoạt động của rạp chiếu phim, từ đặc điểm của khách hàng, đến thông tin cụ thể về vé và phim. Trong dự án này, chúng tôi sẽ tiến hành phân tích dữ liệu này theo ba hướng chính, phản ánh ba mối quan tâm chính của doanh nghiệp:

- Phân khúc và định vị khách hàng tiềm năng: Tập trung vào việc phân loại khách hàng dựa vào các đặc điểm nhân khẩu học và hành vi mua vé, nhằm xác định các nhóm khách hàng tiềm năng để tối ưu hóa chiến lược tiếp thị và quảng cáo.

- Bán chéo popcorn: Đây là việc khám phá cơ hội bán chéo, nơi chúng tôi phân tích mối liên kết giữa việc mua vé xem phim và mua popcorn, cũng như xác định các yếu tố có thể thúc đẩy việc mua hàng phụ. Mục đích là để tối đa hóa lợi nhuận từ mỗi khách hàng bằng cách khuyến khích họ mua thêm popcorn và các sản phẩm khác.

Phân tích sẽ giúp chúng tôi đưa ra các kiến nghị chiến lược dựa trên dữ liệu để tối ưu hóa hiệu quả kinh doanh của rạp chiếu phim. Với một bộ dữ liệu đa dạng và phong phú như vậy, chúng tôi tin rằng phân tích của mình sẽ mở ra những hiểu biết sâu sắc và giúp rạp chiếu phim nâng cao hiệu quả kinh doanh.

b) Biến đổi ngày tháng năm sinh (DOB)

```python
# Hàm chuyển đổi Excel serial date thành year
def excel_serial_date_to_year(serial_date):
    excel_epoch = datetime(1899, 12, 30)
    date = excel_epoch + timedelta(days=serial_date)
    return date.year

# Chuyển đổi DOB từ serial date thành year
df['Year_of_Birth'] = df['DOB'].apply(excel_serial_date_to_year)

# Tính toán tuổi xấp xỉ trong 2019
df['Age_in_2019'] = 2019 - df['Year_of_Birth']

# Hiển thị DataFrame để kiểm tra tính toán
print(df[['customerid', 'DOB', 'Year_of_Birth', 'Age_in_2019','job' ]].head())
```

```python
df = df.drop(['DOB', 'Year_of_Birth'], axis = 1)
```

```python
df_representative = df.groupby('customerid')['Age_in_2019'].first().reset_index()
sns.catplot(y='Age_in_2019', data=df_representative, kind='box')
plt.title('Hình 1.1: Phân phối tuổi khách hàng trước khi xử lý')
plt.show
```

Chúng ta có thể thấy, tuổi của khách hàng có nhiều giá trị âm (ngoại lai) do việc nhập dữ liệu bị lỗi đã ảnh hưởng đến phân phối của dữ liệu.

Giá trị ngoại lai xuất hiện không nhiều nên để tránh ảnh hưởng sâu sắc đến quá trình phân tích dữ liệu thì nhóm sẽ chọn cách thay đổi các giá trị đó. Thay đổi bằng cách đưa các giá trị ngoại lai về cận dưới của biểu đồ Box plot.

```python
# Xử lý các giá trị ngoại lai
Q1 = df_representative['Age_in_2019'].quantile(0.25)
Q3 = df_representative['Age_in_2019'].quantile(0.75)
IQR = Q3 - Q1

Lower_whisker = Q1 - 1.5 * IQR
df['Age_in_2019'] = np.where(df['Age_in_2019'] < Lower_whisker, Lower_whisker, df['Age_in_2019'])
df['Age_in_2019'] = round(df['Age_in_2019'])
```

```python
sns.catplot(y='Age_in_2019', data=df, kind='box')
plt.title('Hình 1.2: Phân phối tuổi khách hàng sau khi xử lý')
plt.show
```

Sau khi xử lý dữ liệu, giá trị ngoại lai vẫn xuất hiện ở trên cận trên của biểu đồ Box plot. Nhưng nhóm quyết định không thay đổi để tránh làm mất mát dữ liệu quan trọng.

```python
# Xử lý null của orderid
def create_unique_orderid(row):
    """Hàm để tạo ra một orderid duy nhất dựa trên customerid, date, và time"""
    unique_string = f"{row['customerid']}-{row['date']}-{row['time']}"
    # Sử dụng kĩ thuật hashing để đảm bảo một orderid duy nhất cho mỗi kết hợp
    hashed_orderid = hashlib.md5(unique_string.encode()).hexdigest()[:10].upper()
    return f"WEB-{hashed_orderid}"

# Áp dụng hàm để tạo một orderid duy nhất cho mỗi hàng liên quan
website_tickets['new_orderid'] = website_tickets.apply(create_unique_orderid, axis=1)

# Bây giờ, chúng ta kết hợp những orderid mới này trở lại vào bảng vé gốc
df.loc[website_tickets.index, 'orderid'] = website_tickets['new_orderid']
```

## 2.3. Phân khúc khách hàng

```python
# Hàm phân khúc khách hàng
def customer_segment(rfm_score):
    if rfm_score in [355, 354, 345, 344, 335]:
        return 'Loyal'
    elif rfm_score in [353, 352, 351, 342, 341, 333, 323]:
        return 'Potential Loyalist'
    elif rfm_score in [311]:
        return 'New Customers'
    elif rfm_score in [315, 314, 313]:
        return 'Promising'
    elif rfm_score in [343, 334, 325, 324]:
        return 'Need Attention'
    elif rfm_score in [331, 321, 312]:
        return 'About To Sleep'
    elif rfm_score in [332, 322]:
        return 'Hibernating Customers'
    else:
        return 'Other'
```

```python
# Phân khúc khách hàng
rfm_df['segment'] = rfm_df['rfm_score'].apply(customer_segment)
```

```python
# Vẽ biểu đồ Tree map
segment_counts = rfm_df['segment'].value_counts()
plt.figure(figsize=(15, 5))
squarify.plot(
    sizes=segment_counts.values,
    label=[None]*len(segment_counts.index),
    color=plt.cm.Set3(range(len(segment_counts.index))),
    pad=0,
)
plt.legend(
    labels=segment_counts.index,
    loc='upper left',
    bbox_to_anchor=(1, 1),
    title='Segments',
    ncol=1
)
plt.title('Hình 2.7: Biểu đồ Tree map thể hiện phân khúc khách hàng')
plt.show()
```

### 3.1.3. Doanh thu các tuần trong tháng 5/2023

```python
# Assume df_ticket_new is your actual loaded DataFrame
ticket_data = df_ticket_new

# Ensure 'saledate' is a datetime column
ticket_data['saledate'] = pd.to_datetime(ticket_data['saledate'])

# Filter for May 2019 (or any other specific time frame you are interested in)
ticket_data = ticket_data[(ticket_data['saledate'] >= '2019-05-01') & (ticket_data['saledate'] <= '2019-05-31')]

# Adding necessary columns
ticket_data['day_of_week'] = ticket_data['saledate'].dt.day_name()
ticket_data['week'] = ticket_data['saledate'].dt.isocalendar().week - ticket_data['saledate'].dt.isocalendar().week.min() + 1

# Group by week and day of week
grouped_data = ticket_data.drop_duplicates(subset='orderid').groupby(['week', 'day_of_week'])['total'].sum().unstack(fill_value=0) / 1e6

# Ensure all days of the week are represented
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
grouped_data = grouped_data.reindex(columns=days, fill_value=0)

num_weeks = grouped_data.shape[0]
fig, axes = plt.subplots(num_weeks, 1, figsize=(10, num_weeks * 4), tight_layout=True)

if num_weeks == 1:  # Ensure axes is always iterable
    axes = [axes]

for week, ax in zip(grouped_data.index, axes):
    bars = ax.bar(grouped_data.loc[week].index, grouped_data.loc[week])
    ax.set_title(f'Tuần thứ {week} của tháng 5/2019')
    ax.set_ylabel('Doanh Thu (Triệu VND)')
    ax.set_xlabel('Ngày trong tuần')

    # Display the values on top of the bars
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}M',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.suptitle('Hình 3.2: Doanh thu của từng tuần trong tháng 5/2019')
plt.show()
```

```python
# Convert 'saledate' to datetime and extract the day of the week
# Monday is 0 and Sunday is 6
new_data['saledate'] = pd.to_datetime(new_data['saledate'])
new_data['day_of_week'] = new_data['saledate'].dt.dayofweek

# Removing duplicates based on 'orderid' and calculating total revenue for each day of the week
daily_revenue_by_weekday = new_data.drop_duplicates(subset='orderid').groupby('day_of_week')['total'].sum()/1e6

# Convert the day_of_week from numbers to actual day names for better readability
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_revenue_by_weekday.index = daily_revenue_by_weekday.index.map(lambda x: days[int(x)])  # Convert to integer

# Setting the aesthetics for the plot
sns.set(style="whitegrid")

# Plotting the total daily revenue for each day of the week
plt.figure(figsize=(10, 6))
daily_revenue_by_weekday.plot(kind='bar')
plt.title('Hình 3.3: Tổng doanh thu các ngày trong tuần')
plt.xlabel('Ngày trong tuần')
plt.ylabel('Doanh Thu (Triệu VND)')
plt.xticks(rotation=45)
plt.show()
```

### 3.2.3. Doanh thu theo từng thể loại

```python
# Removing duplicates to ensure each orderid is counted once
unique_data = new_data

# Tách cột 'listed_in' ra thành các cột dạng binary (0 và 1) cho mỗi thể loại
genre_dummies = unique_data['listed_in'].str.get_dummies(sep=', ')

# Kết hợp dữ liệu genre_dummies với dữ liệu gốc
data_with_genres = pd.concat([unique_data, genre_dummies], axis=1)
```

```python
# Removing duplicates to ensure each orderid is counted once
data_with_genres = data_with_genres.drop_duplicates(subset=['orderid'])

# Tính doanh thu cho mỗi thể loại
# Lọc ra các cột thể loại phim dạng binary và cột 'total'
binary_columns = data_with_genres.columns.str.match(r'(Animation|Action-Adventure|Comedy|Crime|Drama|Family|Fantasy|Horror|Mystery|Romance|Science Fiction|Thriller)$')
binary_genre_data = data_with_genres.loc[:, binary_columns]
revenue_by_genre = binary_genre_data.multiply(data_with_genres['total'], axis='index')

# Tính tổng doanh thu cho mỗi thể loại
revenue_by_genre_summed = revenue_by_genre.sum()/1e6
# Sắp xếp doanh thu theo từng thể loại phim từ cao xuống thấp
revenue_by_genre_summed_sorted = revenue_by_genre_summed.sort_values(ascending=False)

# Vẽ biểu đồ biểu diễn doanh thu theo từng thể loại phim (đã sắp xếp)
plt.figure(figsize=(14, 7))
ax = revenue_by_genre_summed_sorted.plot(kind='bar')
plt.title('Hình 3.10: Doanh thu theo từng thể loại phim')
plt.xlabel('Thể loại phim')
plt.ylabel('Tổng doanh thu')
plt.xticks(rotation=45)

# Thêm chú thích số liệu lên trên từng cột
for i, v in enumerate(revenue_by_genre_summed_sorted):
    ax.text(i, v, f'{v:.2f}M', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```


**Xu hướng thể loại phim:**
* Kinh dị (Horror): Có tổng doanh thu cao nhất,cho thấy rằng thể loại này đang rất phổ biến. Thể loại kinh dị thường hợp với lứa tuổi từ thiếu niên đến người lớn và thường được tìm kiếm bởi những người yêu thích cảm giác mạnh và kích thích tinh thần.

* Hành động-Phiêu lưu (Action-Adventure): Cũng có doanh thu cao, thể hiện sự ưa chuộng của khán giả với những bộ phim mang tính chất giải trí cao, hành động mãnh liệt và câu chuyện phiêu lưu hấp dẫn. Phim hành động thường hướng đến đối tượng khán giả rộng từ trẻ em (có sự giám sát của người lớn) đến người lớn.

* Fantasy và Thriller: Cũng có doanh thu tốt, thể hiện sự yêu thích đối với những bộ phim mang yếu tố thần thoại, siêu nhiên hoặc những câu chuyện căng thẳng, kích thích trí óc.

* Khoa học viễn tưởng (Science Fiction): Với doanh thu đứng ở mức trung bình, phim khoa học viễn tưởng thường thu hút khán giả trẻ tuổi và trung niên quan tâm đến khoa học, công nghệ và tương lai.

* Hài (Comedy): Phim hài có sức hấp dẫn rộng rãi, từ trẻ em đến người già, tuy nhiên, doanh thu không cao bằng các thể loại đầu bảng, có thể do sự cạnh tranh mạnh mẽ từ các thể loại khác.

* Gia đình (Family): Đứng thấp trong danh sách về doanh thu, nhưng thường hấp dẫn đối với gia đình có con nhỏ và có khả năng tăng doanh thu vào các dịp lễ hoặc kỳ nghỉ.


```python
# Lấy danh sách các thể loại phim
genres = ['Animation', 'Action-Adventure', 'Comedy', 'Crime', 'Drama', 'Family', 'Fantasy', 'Horror', 'Mystery', 'Romance', 'Science Fiction', 'Thriller']

# Loại bỏ các hàng bị trùng lặp dựa trên cột 'title'
data_with_genres = data_with_genres.drop_duplicates(subset=['film'])

# Duyệt qua mỗi thể loại và in ra danh sách các bộ phim thuộc thể loại đó
for genre in genres:
    movies_in_genre = data_with_genres[data_with_genres[genre] == 1]['film']
    print(f"Thể loại '{genre}':")
    for movie in movies_in_genre:
        print(f"- {movie}")
    print("\n")
```

### 4.2.7. Khám phá slot type VS popcorn

### 4.2.11. Mua vé trước có ảnh hưởng đến quyết định mua bắp?

```python
# Để thực hiện EDA trên 'time_before_days_hours', chúng ta cần chuyển đổi dữ liệu này trở lại thành số giờ
# Tính toán thời gian mua vé trước khi xem phim (time_before) theo giờ
df3['time_before'] = (pd.to_datetime(df3['date'].astype(str) + ' ' + df3['time'].astype(str)) -
                      pd.to_datetime(df3['sale day'].astype(str) + ' ' + df3['sale time'].astype(str))) / pd.Timedelta(hours=1)

# Phân loại 'time_before' thành các khoảng thời gian
def classify_time_before(hours):
    if hours <= 1:
        return '0-1 hour'
    elif hours <= 2:
        return '1-2 hours'
    elif hours <= 3:
        return '2-3 hours'
    elif hours <= 4:
        return '3-4 hours'
    elif hours <= 5:
        return '4-5 hours'
    else:
        return '5+ hours'

# Áp dụng phân loại
df3['time_before_category'] = df3['time_before'].apply(classify_time_before)

# Nhóm theo cột mới 'time_before_category' và tính toán tỷ lệ mua bỏ bỏ trung bình
time_before_popcorn_rate = df3.groupby('time_before_category')['popcorn'].mean() * 100

# Chuẩn bị dữ liệu cho việc trực quan hóa
time_before_popcorn_rate_df = time_before_popcorn_rate.reset_index()

# Sắp xếp theo tỷ lệ mua bắp trung bình
time_before_popcorn_rate_df = time_before_popcorn_rate_df.sort_values('popcorn', ascending=False)

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.bar(time_before_popcorn_rate_df['time_before_category'], time_before_popcorn_rate_df['popcorn'], color='lightblue')
plt.xlabel('Thời Gian Mua Trước (giờ)')
plt.ylabel('Tỷ Lệ Mua Bắp Rang (%)')
plt.title('Hình 4.11: Thời Gian Mua Vé Trước Khi Xem Phim So Với Tỷ Lệ Mua Bắp Rang')
plt.xticks(rotation=45)
plt.show()
```

Biểu đồ cột thể hiện tỷ lệ mua bắp rang dựa trên các khoảng thời gian khách hàng mua vé trước khi xem phim. Tỷ lệ mua bắp rang cao nhất ở khoảng thời gian mua vé từ 3-4 giờ trước khi xem phim, và tỷ lệ giảm dần khi thời gian mua vé xa hơn so với thời điểm xem phim.

**Insights:**

- **Mua sớm, mua bắp rang nhiều**: Khách hàng có xu hướng mua bắp rang cao hơn khi họ mua vé từ 3-4 giờ trước buổi chiếu. Điều này có thể phản ánh kế hoạch xem phim được chuẩn bị trước và quyết định mua đồ ăn nhẹ cũng được lên kế hoạch từ trước.
- **Mua gần giờ chiếu, mua bắp rang ít**: Tỷ lệ mua bắp rang giảm ở những khách hàng mua vé gần với thời gian chiếu phim, có thể do họ ít có thời gian để suy nghĩ hoặc quyết định mua thêm đồ ăn nhẹ.
- **Cơ hội tối ưu**: Rạp chiếu phim có thể cân nhắc cung cấp thông tin và quảng cáo bắp rang mạnh mẽ hơn vào khoảng thời gian 3-4 giờ trước buổi chiếu để tận dụng xu hướng mua hàng này. Đồng thời, cũng nên cải thiện trải nghiệm mua hàng vào những khoảng thời gian khác để khuyến khích mua bắp rang.

Biểu đồ trên thể hiện tổng doanh thu từ vé xem phim cho mỗi thể loại phim. Sự phân bố của tổng doanh thu theo thể loại cho thấy một số thể loại thu hút được nhiều khán giả hơn và do đó tạo ra doanh thu cao hơn.
